```python
##trump data
Trump_tweets = pd.read_csv("/Users/zhangmingshan/Desktop/2018 Autumn/Data Analytics/Project/twitter raw data/25073877_Donald J. Trump_tweets.csv",header=None)
Trump_time=Trump_tweets[[0,1]]
Trump_time[1]=Trump_time[1].apply(lambda x:time.strptime(x,"%Y-%m-%d %H:%M:%S"))
Trump_time=Trump_time.sort_values(1)
Trump_time.columns=['ID','Date']
Trump_time['Date_int']=Trump_time['Date'].apply(lambda x:time.mktime(x))
Trump_time
```

```python
def find_index(x):
    a=list(np.where((df_close_new['Time_int']-60<x) & (x<60+df_close_new['Time_int'])))[0]
    if a.size ==0:
           return 'NaN'
    else:
           return a[:1]
```

```python
##seach t1+interval price
def find_interval_later(t1,unit):
    interval=unit*60
    if find_index(t1+interval)=='NaN':
        if find_index(t1+interval+60)=='NaN':
            if find_index(t1+interval-60)=='NaN':
                return 'NaN'
            else:
                return (df_close_new['Close'].iloc[find_index(t1+interval-60)].values[0])
        else:
            return (df_close_new['Close'].iloc[find_index(t1+interval+60)].values[0])
    else:
        return (df_close_new['Close'].iloc[find_index(t1+interval)].values[0])
```

```python
#find current_time
def find_tweet_price(t1):
    if find_index(t1)=='NaN':
        if find_index(t1-60)=='NaN':
            return 'NaN'
        else:
            return (df_close_new['Close'].iloc[find_index(t1-60)].values[0])
    else:
        return (df_close_new['Close'].iloc[find_index(t1)].values[0])
```

```python
 Trump_time['tweet_price']=Trump_time['Date_int'].apply(lambda x:find_tweet_price(x))
```

```python
for unit in (1,3,5,10,15,30):
    Trump_time['_%s_interval_price'%unit]=Trump_time['Date_int'].apply(lambda x:find_interval_later(x,unit))
    
```

```python
Trump_time
```

```python
def get_tag(x,unit):
    if x['tweet_price']=='NaN' or x['_%s_interval_price'%(unit)]=='NaN':
        return ('NaN')
    elif (x['_%s_interval_price'%(unit)])/(x['tweet_price']) >1:
        return ('1')
    elif (x['_%s_interval_price'%(unit)])/(x['tweet_price']) <1:
        return ('-1')
    elif (x['_%s_interval_price'%(unit)])/(x['tweet_price']) ==1:
        return ('0')
    else:
        return ('NaN')
```

```python
for unit in (1,3,5,10,15,30):
    Trump_time['_%s_interval_tag'%unit]=Trump_time.apply(lambda x:get_tag(x,unit),axis=1)
```
